```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a simple dataset
data = pd.DataFrame({
    'Year': [2020, 2021, 2022, 2023],
    'Sales': [100, 150, 200, 300]
})

# Plot a line graph
plt.plot(data['Year'], data['Sales'], marker='o')
plt.title("Yearly Sales Growth")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)   # optional, adds grid lines
plt.show()
```
